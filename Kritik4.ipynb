```python
#Q4 f(x) = x^2
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point): #define the  function (from the instructions)
    def deriv(f, base_point):
        return (f(base_point + 10**(-5)) - f(base_point - 10**(-5))) / (2 * 10**(-5)) #from the isntructions

    x_coords = [initial_point] #from the isntructions
    y_coords = [f(initial_point)]

    for _ in range(100): #for loop to find the minima by moving in the direction with the steepest slope.
        gradient = deriv(f, x_coords[-1]) #taking the derivative of f(x)
        update = -learning_rate * gradient #update after each point, since they have different slopes.
        x_coords.append(x_coords[-1] + update)
        y_coords.append(f(x_coords[-1])) # we do not want to plot more points than necessary

    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 10000) #from the isntructions
    function_range = [f(i) for i in plot_range]

#plotting the graph (for visual sake). Note that this may not be necessary but i find it helps me visualize wheteher a min or max will occur.
    plt.plot(plot_range, function_range, label='f(x)') #plot the function y = x^2
    plt.plot(x_coords, y_coords, 'o-', label='Gradient Descent Path') # the orange line, shows teh "cobwebbing" explored in lectures
    plt.xlabel('x') #self explanatory
    plt.ylabel('f(x)') #self explanatory
    plt.legend() #self explanatory
    plt.show() #self explanatory

    return round(x_coords[-1], 3), round(y_coords[-1], 3) #from the isntructions

#  use lambda to consider the  function we were given i.e., y = x^2, i have experimentally determined an appropriate learning_rate and initial_point
result = gradient_descent(lambda x: x**2, 0.1, 3.0)
print("Minimum point:", result) #self explanatory

#The notes above apply to all of the following code
```
